Gráfico das 10 Publicadoras que mais venderam cópias dos jogos

---

```python
publicadoras_com_mais_receita = games.groupby('Publisher')['Global_Sales'].sum()


# Selecionar as empresas de produção com maior revenue
top_10_publicadoras = publicadoras_com_mais_receita.nlargest(10)

display(top_10_publicadoras)


plt.style.use('seaborn-v0_8')

# Criar o gráfico de barras para as empresas com maior revenue
plt.figure(figsize=(12, 6))
plt.bar(top_10_publicadoras.index, top_10_publicadoras.values, color='purple')

# Geração de gráfico onde o x é "Companhia de Produção", e y é "Revenue(em bilhões)"
plt.xlabel('Produtora')
plt.ylabel('Global Sales')
plt.title('Top 10 Publicadoras com Maior Número de vendas')

# Ajustar tamanho da fonte e rotação dos rótulos do eixo x
plt.xticks(rotation=45, ha='right', fontsize=12)



plt.show()
```

Gráficos que demonstram como a venda do  número de cópias dos jogos se comportou ao longo dos anos em cada região

```python

lucros_japao= games.groupby('Year_of_Release')['JP_Sales'].sum()



lucros_japao= lucros_japao.dropna()




anos_lancamento_unicos=lucros_japao.index.unique()
lucros_japao_unicos= lucros_japao.index.unique()









plt.style.use('seaborn')



plt.barh( lucros_japao.index,lucros_japao.values/10**10, color='blue')
plt.title('Vendas no Japão')
plt.xlabel('Número de cópias vendidas')
plt.ylabel('Ano de lançamento')
plt.xticks( rotation=45)
plt.tight_layout()
plt.show()
```

```python
lucros_europa= games[['Year_of_Release', 'EU_Sales']]
lucros_europa= lucros_europa.dropna()
lucros_europa = lucros_europa.groupby(by = 'Year_of_Release').sum().reset_index()
display(lucros_europa)

anos_lancamento_unicos_europa= lucros_europa.index.unique()
plt.style.use('seaborn')

#plt.barh(lucros_northamerica.index, lucros_northamerica.index/10**12, color='green', alpha=0.5)
plt.bar(x = lucros_europa['Year_of_Release'], height= lucros_europa['EU_Sales'])
plt.title('Lucros na Europa')
plt.xlabel('Ano de lançamento')
plt.ylabel('Lucro')
plt.xticks(rotation=45)

# Ajustar o espaçamento entre os subplots para evitar sobreposição
plt.tight_layout()

# Exibir o gráfico
plt.show()
```

```python
lucros_northamerica= games[['Year_of_Release', 'NA_Sales']]
lucros_northamerica= lucros_northamerica.dropna()
lucros_northamerica = lucros_northamerica.groupby(by = 'Year_of_Release').sum().reset_index()
display(lucros_northamerica)

anos_lancamento_unicos_northamerica= lucros_northamerica.index.unique()
plt.style.use('seaborn')

#plt.barh(lucros_northamerica.index, lucros_northamerica.index/10**12, color='green', alpha=0.5)
plt.bar(x = lucros_northamerica['Year_of_Release'], height= lucros_northamerica['NA_Sales'])
plt.title('Lucros na América do Norte')
plt.xlabel('Ano de lançamento')
plt.ylabel('Lucro')
plt.xticks(rotation=45)

# Ajustar o espaçamento entre os subplots para evitar sobreposição
plt.tight_layout()

# Exibir o gráfico
plt.show()
```

### Gráfico das vendas globais utilizando como parâmetro de comparação as desenvolvedoras





---



```python
desenvolvedoras_com_mais_receita= games.groupby('Developer')['Global_Sales'].sum()

top_15_desenvolvedoras= desenvolvedoras_com_mais_receita.nlargest(15)

plt.figure(figsize=(12,6))


plt.xlabel('Desenvolvedores')
plt.ylabel('Receita')
plt.title('Relação entre as vendas e as desenvolvedoras')
plt.bar(top_15_desenvolvedoras.index, top_15_desenvolvedoras.values/10**11, color='red')
plt.xticks(rotation=45, ha='right', fontsize=12)





plt.style.use('seaborn-v0_8')
plt.tight_layout()
plt.show()
```

Gráfico de pizza com os 6 jogos mais bem avaliados pelos críticos

```python
games['Critic_Score'] = pd.to_numeric(games['Critic_Score'], errors='coerce')
games = games.drop_duplicates(subset='Name', keep='first')
top_games = games.nlargest(6, 'Critic_Score')
top_games_names=top_games['Name'].values
top_games_scores=top_games['Critic_Score'].values


plt.figure(figsize=(8,8))
plt.pie(top_games_scores, labels=top_games_names, startangle=140 )
plt.axis('equal')
plt.title('6 jogos mais bem pontuados pelos usuários')
plt.show()
```
